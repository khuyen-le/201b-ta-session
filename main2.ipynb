```python
# Part 2: Debugging

# Debugging the intentional errors from Part 1.

# Fixing Syntax Error
def calculate_square_fixed(x):
    return x * x

# Fixing Logical Error
def sum_of_list_fixed(numbers):
    total = 0
    for number in numbers:
        total += number  # Corrected logic
    return total

# Fixing Type Error
def greet_fixed(name, age):
    return "Hello, " + name + "! You are " + str(age) + " years old."

# Fixing Name Error
def plot_data_fixed(data):
    plt.plot(data)
    plt.title("Data Plot")
    plt.show()  # Removed 'fig'

# Fixing Index Error
def get_last_element_fixed(lst):
    return lst[len(lst) - 1]  # Corrected index

# Fixing ZeroDivisionError
def divide_numbers_fixed(a, b):
    if b == 0:
        return "Cannot divide by zero"
    return a / b

# Fixing ValueError
def convert_to_int_fixed(value):
    try:
        return int(value)
    except ValueError:
        return "Cannot convert to integer"

# Fixing KeyError
def get_value_fixed(dictionary, key):
    return dictionary.get(key, "Key not found")

# Fixing AttributeError
def call_method_fixed(obj):
    if hasattr(obj, 'non_existent_method'):
        return obj.non_existent_method()
    else:
        return "Method does not exist"

# Fixing ImportError
# No fix needed, just ensure the module exists before importing

# Part 3: Data Visualization

# Let's create some visualizations using NumPy, Matplotlib, and Seaborn.

# Generating sample data
np.random.seed(0)
data = np.random.randn(100)

# Matplotlib Example 1: Histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=20, alpha=0.7, color='blue')
plt.title('Histogram of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Matplotlib Example 2: Line Plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sine Wave')
plt.title('Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()

# Seaborn Example 1: Boxplot
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title('Boxplot of Total Bill by Day')
plt.show()

# Seaborn Example 2: Scatter Plot
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()

# Seaborn Example 3: Heatmap
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot("month", "year", "passengers")
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Passenger Flights')
plt.show()

# Part 4: Bootstrapping

# Bootstrapping is a statistical method that involves sampling with replacement.
# Let's demonstrate bootstrapping with NumPy.

def bootstrap_mean(data, num_samples=1000):
    means = []
    for _ in range(num_samples):
        sample = np.random.choice(data, size=len(data), replace=True)
        means.append(np.mean(sample))
    return means

# Bootstrapping the mean of the sample data
means = bootstrap_mean(data)

# Plotting the distribution of bootstrapped means
plt.figure(figsize=(10, 6))
sns.histplot(means, kde=True, color='green')
plt.title('Distribution of Bootstrapped Means')
plt.xlabel('Mean')
plt.ylabel('Frequency')
plt.show()

# Bootstrapping Example 2: Confidence Interval
def bootstrap_confidence_interval(data, num_samples=1000, confidence_level=0.95):
    means = bootstrap_mean(data, num_samples)
    lower_bound = np.percentile(means, (1 - confidence_level) / 2 * 100)
    upper_bound = np.percentile(means, (1 + confidence_level) / 2 * 100)
    return lower_bound, upper_bound

lower_bound, upper_bound = bootstrap_confidence_interval(data)
print(f"Confidence Interval: ({lower_bound}, {upper_bound})")

# Bootstrapping Example 3: Comparing Means
data1 = np.random.normal(loc=50, scale=10, size=100)
data2 = np.random.normal(loc=55, scale=10, size=100)

def bootstrap_compare_means(data1, data2, num_samples=1000):
    mean_diff = []
    for _ in range(num_samples):
        sample1 = np.random.choice(data1, size=len(data1), replace=True)
        sample2 = np.random.choice(data2, size=len(data2), replace=True)
        mean_diff.append(np.mean(sample1) - np.mean(sample2))
    return mean_diff

mean_diff = bootstrap_compare_means(data1, data2)

plt.figure(figsize=(10, 6))
sns.histplot(mean_diff, kde=True, color='purple')
plt.title('Distribution of Bootstrapped Mean Differences')
plt.xlabel('Mean Difference')
plt.ylabel('Frequency')
plt.show()

# Done
```
